```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
DATA =  pd.read_excel("Ina.xlsx")
data = DATA[['Country Name', 'Years',
       'Primary completion rate, total (% of relevant age group)',
       'GDP (current US$)', 'Inflation, consumer prices (annual %)',
       'GDP growth (annual %)',"Birth rate, crude (per 1,000 people)","Urban population growth (annual %)"]]
```

```python
for col in data.columns:
    for i,line in zip(data.index,data[col]) :
        if line == '..' :
            data = data.drop(i)
            #print(line)
```

```python
benin = data[data["Country Name"]=="Benin"]
benin.head()
```

```python
benin.columns
```

```python
fig , axes = plt.subplots(nrows=3,ncols=2,figsize=(14,10),dpi=300)

snb.color_palette("flare")

snb.lineplot(data=benin,x='Years',y='Primary completion rate, total (% of relevant age group)',ax=axes[0,0],color='darkgreen')
axes[0,0].set_title('Primary completion rate, total (% of relevant age group)',color='darkgreen')

snb.lineplot(data=benin,x='Years',y='GDP (current US$)',ax=axes[1,0])
axes[1,0].set_title('GDP (current US$)')

snb.lineplot(data=benin,x='Years',y='Inflation, consumer prices (annual %)',ax=axes[2,0])
axes[2,0].set_title('Inflation, consumer prices (annual %)')

snb.lineplot(data=benin,x='Years',y= 'GDP growth (annual %)',ax=axes[0,1])
axes[0,1].set_title('GDP growth (annual %)')

snb.lineplot(data=benin,x='Years',y='Birth rate, crude (per 1,000 people)',ax=axes[1,1])
axes[1,1].set_title('Birth rate, crude (per 1,000 people)')

snb.lineplot(data=benin,x='Years',y='Urban population growth (annual %)',ax=axes[2,1],color='darkred')
axes[2,1].set_title('Urban population growth (annual %)',color='darkred')

fig.suptitle("Representation deux à deux des variables pour le Benin")
plt.tight_layout()
```
