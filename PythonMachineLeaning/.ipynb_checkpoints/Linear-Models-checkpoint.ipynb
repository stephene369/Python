# Simple lenear regression

```python
snb.set_style("whitegrid")
snb.lmplot(x='total_bill' , y='tip' , data=tips , hue='sex' , 
           markers=['o','v'] , scatter_kws= {'s':100} , palette='plasma')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test,y_train , y_test = train_test_split(np.array(tips['total_bill']).reshape(-1,1),
                                                    np.array(tips['tip'])
                                                    , random_state=42)

print("X_train shape {}".format(X_train.shape))
print("x_test shape {}\n".format(X_test.shape))
print("y_train shape {}".format(y_train.shape))
print("y_test shape {}".format(y_test.shape))
```

```python
lr = LinearRegression().fit(X_train,y_train)
print("Lr Coefficients: {}".format(lr.coef_))
print("\nLr.intercept : {}".format(lr.intercept_))
```

```python
print("Score d'entrainement: {}".format(lr.score(X_train,y_train)) , color='green')
print("\nScore du test: {}".format(lr.score(X_test,y_test)) , color='green')
```

# Multiple Lenear Regression

```python
snb.set_style("whitegrid")
snb.lmplot(x='total_bill' , y='tip' , data=tips , hue='smoker' , col='sex',
           scatter_kws= {'s':100} , palette='plasma')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test,y_train , y_test = train_test_split(np.array(tips[['total_bill','size','price_per_person']]),
                                                    np.array(tips['tip'])
                                                    , random_state=42)

print("X_train shape {}".format(X_train.shape))
print("x_test shape {}\n".format(X_test.shape))
print("y_train shape {}".format(y_train.shape))
print("y_test shape {}".format(y_test.shape))
```

```python
lr = LinearRegression().fit(X_train,y_train)
print("Lr Coefficients: {}".format(lr.coef_))
print("\nLr.intercept : {}".format(lr.intercept_))
```

```python
print("Score d'entrainement: {}".format(lr.score(X_train,y_train)) , color='green')
print("\nScore du test: {}".format(lr.score(X_test,y_test)) , color='green')
```
