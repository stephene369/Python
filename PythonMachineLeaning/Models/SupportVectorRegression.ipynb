```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.svm import SVR 
```

```python
X = np.sort( 5*np.random.rand(40 , 1) , axis = 0 )
y = np.sin(X).ravel()
```

```python
# adding dome noise values
y[::5] += 3*(0.5 - np.random.rand(8) )
```

### Model on the sine Curve data using Lineat Kernel

```python
# create SVR models with kernel 
svr = SVR(kernel='linear')
```

```python
# train the models
svr.fit(X , y)
```

```python
# prediction on the data 
y_pred = svr.predict(X)
print(y_pred)
```

```python
# plot the predicted values againdt true values 
plt.scatter( X  , y)
```

```python
plt.scatter(X , y , color="darkorange", label='data')
plt.plot(X , y_pred, color = 'cornflowerblue', label='prediction' )
plt.legend()
plt.show()
```

### Model on the sine curve data using Polynomial Kernel 

```python
svr_ = SVR(kernel="poly")
svr_.fit(X , y)
```

```python
y_pred_ = svr_.predict(X) 
```

```python
plt.scatter(X , y , color='darkorange' , label="data")
plt.plot(X , y_pred_ , color='cornflowerblue',label='prediction')
plt.legend()
plt.show()
```

### Model on the sine Curve data using RBF Kernel 

```python
svr__ = SVR(kernel='rbf') 
svr__.fit(X,y)
```

```python
y_pred__ = svr__.predict(X) 
```

```python
plt.scatter(X , y , color='darkorange' , label="data")
plt.plot(X , y_pred__ , color='cornflowerblue',label='prediction')
plt.legend()
plt.show()
```
