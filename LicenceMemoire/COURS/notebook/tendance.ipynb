## Services Publique

```python
import pandas as pd
import numpy as np
BRVM = pd.read_csv("data/BRVM-Public-Services.csv",index_col="Date")
import matplotlib.pyplot as plt
debut = -730-365
brvm = BRVM.iloc[debut:-365]
```

```python
resistance = [[],[]]
support = [[],[]]
j=0
for i in range(0, len(brvm) ,100) :
    try :
        
        max_ = brvm[j:i]['Close'].max()
        min_ = brvm[j:i]['Close'].min()

        date = brvm[j:i][brvm[j:i]['Close']==max_].index[0]
        date_ = brvm[j:i][brvm[j:i]['Close']==min_].index[0]

        resistance[0].append(date)
        resistance[1].append(max_)

        support[0].append(date_)
        support[1].append(min_)

        j=i

    except Exception as e: print(e)
#####
piqueP = [[],[]]
piqueN = [[],[]]
j=0
for i in range(0, len(brvm) ,50) :
    try :
        max_ = brvm[j:i]['Close'].max()
        min_ = brvm[j:i]['Close'].min()

        date = brvm[j:i][brvm[j:i]['Close']==max_].index[0]
        date_ = brvm[j:i][brvm[j:i]['Close']==min_].index[0]

        piqueP[0].append(date)
        piqueP[1].append(max_)

        piqueN[0].append(date_)
        piqueN[1].append(min_)

        j=i
    except Exception as e: print(e)
```

```python
print(resistance[0],resistance[1])
```

```python
plt.figure(figsize=(19,11),dpi=150)
plt.plot(brvm.index , brvm["Close"])
position = []
for i in range(0 , len(brvm) , 100) : position.append(i)
labels = [brvm.iloc[int(i)].name for i in position]
plt.xticks(position,labels,rotation=45)

plt.plot(resistance[0] , resistance[1] , marker='^',color='red',markersize=6,label = "Resistance",linewidth=0.5)
plt.line(support[0] , support[1] , marker='v',color='green',markersize=6,label = "Support",linewidth=0.5)

plt.plot(piqueP[0] , piqueP[1] , marker='^',color='red',markersize=3,linewidth=0)
plt.plot(piqueN[0] , piqueN[1] , marker='^',color='green',markersize=3,linewidth=0)

plt.xlabel("Date")
plt.ylabel("Prix de clture")
plt.title("Tendance et niveaux de resistance et support de l'indice BRVM-Agriculture")
plt.legend()
#plt.savefig(r"C:\Users\steph\OneDrive\Documents\COURS\latex\img\agri_tendance.jpg")
```
